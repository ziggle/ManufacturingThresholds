```python
import numpy as np
import matplotlib.pyplot as plt
import math
t = np.linspace(0,2,100)
f1= lambda x: math.exp(-(x/3)**3)
f2= lambda x: math.exp(-(x/4)**2)
f3= lambda x: math.exp(-(x/3)**3+(x/4)**2)
vf1 =np.vectorize(f1)
vf2 =np.vectorize(f2)
vf3 =np.vectorize(f3)


y  = vf1(t)
y2 = vf2(t)
y3 = vf3(t)*y2

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.plot(t,y, 'r')
plt.plot(t,y2,'b')
plt.plot(t,y3,'g--')
plt.show()
```
